```python
#1.IMPORTAR LIBRERIAS:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from IPython.display import display

#2.CARGA DE DATOS:

urlCSV = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRV4vMyWjcQBfDm39bHNlzcnSp-UiAn_01otTFSxBgN7P9IRcw7hM9tq3XBHbd8TkJ-NECV6lQFdHY2/pub?gid=1967932107&single=true&output=csv'

data_frame = pd.read_csv(urlCSV)

print("PRIMER FILA DATAFRAME \n")
display(data_frame.head(1))

print("\n El dataset se trata de las ventas (1000) de un supermercado durante el 2018-2019 y contiene 17 columnas en total las cuales estan distribuidas 8 como stings 7  numericas 2 tipo Date Time \n")
print("Las variables son las siguientes: \n ID de factura: Tipo String \n Sucursal: Tipo String \n Ciudad: Tipo String \n Tipo de cliente: Tipo String \n Genero (clientes): Tipo String \n Línea de producto: Tipo String \n Precio unitario: Tipo float \n Cantidad: Tipo int \n Impuesto 5%: Tipo float \n Total: Tipo float \n Fecha: Tipo Datatime \n Hora: Tipo Datatime \n Pago: Tipo string Costo: Tipo float \n Porcentaje de margen bruto: Tipo String (aunque deberia ser float) \n Ingresos brutos: Tipo float \n Clasificación o Rating: Tipo Float \n ")
print("\n VALORES NULOS \n")
nulos = data_frame.isnull().sum()
print(nulos)

print("NO HUBO DATOS NULOS; ATIPICOS O AUSENTES PARA ELIMINAR \n")

#ELIMINACION:
print("DATAFRAME LIMPIO\n")
data_frame_limpio = data_frame.dropna()
#data_frame_limpio['gross margin percentage']=data_frame_limpio['gross margin percentage'].astype(float) , Nose porque no se puede convertir esta columna de string a float por eso la elimine mas abajo
data_frame_limpio = data_frame_limpio.drop('gross margin percentage', axis=1)
print(data_frame_limpio)

#4. GRAFICA DATAFRAME ANTES DEL PREPROCESAMIENTO
print("\nGRAFICO SUCIO: DEL TOTAL DE VENTAS VS PRECIO UNITTARIO DE LOS PRODUCTOS \n")
plt.figure(figsize=(10, 6))
plt.scatter(data_frame['Total'], data_frame['Unit price'],)
plt.xlabel('Total de ventas')
plt.ylabel('Precio Unitario')
plt.title('ANTES DEL PREPROCESAMIENTO')
plt.show()

print("\n\n")

#5. GRAFICA DATAFRAME DESPUES DEL PREPROCESAMIENTO
print("GRAFICO LIMPIO: DEL TOTAL DE VENTAS VS PRECIO UNITTARIO DE LOS PRODUCTOS \n")
plt.figure(figsize=(10, 6))
plt.scatter(data_frame_limpio['Total'], data_frame_limpio['Unit price'])
plt.xlabel('Total de ventas')
plt.ylabel('Precio Unitario')
plt.title('DESPUES DEL PREPROCESAMIENTO')
plt.show()

print("\n COMO NO HUBO DATOS NULOS QUE ELIMINAR, LOS GRAFICOS SALIERON IGUALES \n")

#ANÁLISIS DE FRECUENCIA
data_frame['Payment'].value_counts().plot(kind='bar')
plt.xlabel('Pago')
plt.ylabel('Frecuencia')
plt.title("Análisis de frecuencia por forma de pago")
plt.show()

print("\n PODEMOS OBSERVAR DEL ANTERIROR GRAFICO QUE LOS CLIENTES PREFIEREN ABONAR SUS COMPRAS CON DINERO EN EFECTIVO O BILLETERA VIRTUAL, SIN EMPBARGO LOS PAGOS CON TARJETAS DE CREDITO ES MENOR PERO NO POR MUCHA DIFERENCIA \n")

#ANÁLISIS UNIVARIADO
plt.figure(figsize = (12,10))
plt.subplot(2, 2, 1)
data_frame['Unit price'].plot(kind='hist')
plt.xlabel('Precio Unitario')
plt.title('Histograma de Precios unitarios de los productos')

plt.subplot(2, 2, 2)
data_frame['Quantity'].plot(kind='hist')
plt.xlabel('Cantidad')
plt.title('Histograma de cantidad por ventas')

plt.subplot(2, 2, 3)
data_frame['Total'].plot(kind='hist')
plt.xlabel('Total por ventas')
plt.title('Histograma del Total por cada venta')

plt.subplot(2, 2, 4)
data_frame['cogs'].plot(kind='hist')
plt.xlabel('Costo de los productos vendidos')
plt.title('Histograma de los Costos de los productos vendidos')
plt.show()

print("\n EN EL ANALISIS UNIVARIADO DE PRECIO UNITARIO PODEMOS OBSERVAR QUE LOS NO LES IMPORTARIA MUCHO EL PRECIO DE LOS PRODUCTOS YA QUE TIENEN UNA FRECUENCIA MUY SIMILAR EN LOS DISTINTOS RANGOS Y EL QUE MAS DESTACA ES EL MAS CARO \n")
print("\n EN EL ANALISIS UNIVARIADO DE LAS CANTIDAD QUE SE VENDEN PAR CADA VENTA DE UN PRODUCTO  DESTACAN POR MAYOR FRECUENCIA LAS CATINDADES 1 Y 10 Y EL RESTO ES MUY PAREJO \n")
print("\n EN EL ANALISIS UNIVARIADO DEL TOTAL POR VENTA (PRECIO UNITARIO x CANTIDAD) PODEMOS OBSERVAR  QUE LA MAYORIA DE LAS VENTAS SON ENTRE 50 Y 200 DOLARES Y VA DECRECIENDO NOTABLEMENTE HASTA LLEGAR A LAS MENOS FRECUENTES QUE SON LAS VENTAS DE 1000 DOLARES \n")
print("\n EN EL ANALISIS UNIVARIADO DEL COSTO DE LOS PRODUCTOS VENDIDOS OBSERVAMOS QUE EL GRAFICO ES SIMILAR AL DE VENTAS YA QUE EL MARGEN DE GANANCIA ES SOLO DEL 5% Y SE APLICA DE LA MISMA FORMA A TODOS LOS PRODUCTOS \n")

#ANALISIS MULTIVARIADO:

correlacion_matrix= data_frame_limpio.corr(numeric_only=True)

#Graficamos CORRELACION:

plt.figure(figsize=(12,10))
plt.imshow(correlacion_matrix, cmap='coolwarm', interpolation= "nearest")
plt.colorbar()
plt.xticks(range(len(correlacion_matrix)), correlacion_matrix.columns, rotation=45)
plt.yticks(range(len(correlacion_matrix)), correlacion_matrix.columns, rotation=45)
plt.title(' MATRIZ DE CORRELACION ')
plt.show()

print("\n LAS VARIABLES TOTAL POR VENTAS, COSTO DE PRODUCTOS POR VENTAS, IMPUESTO DEL 5% Y INGRESOS BRUTOS ESTAN FUERTEMENTE CORRELACIONADAS ENTRE SI, Y RATING NO PRESENTA NINGUNA CORRELACION CON EL RESTO DE VARIABLES \n")

#Preprocesamiento de DATOS: Hacer ESTANDARIZACION
features_name= ['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross income', 'Rating']
scaler= StandardScaler()
ventas_mercado_scaled= scaler.fit_transform(data_frame_limpio[features_name])

#APLICAR PCA: Aplicar el calculo de los componentes principales
pca= PCA()
pca.fit(ventas_mercado_scaled)

#Aplicamos FUNCIONES para ANALIZAR: En este caso aplicaremos VARIANZA
explicacion_varianza= pca.explained_variance_ratio_
acumulacion_explicacion_varianza= np.cumsum(explicacion_varianza)

#Graficar la VARIANZA EXPLICADA COMULATIVA

plt.figure(figsize=(12,6))
plt.plot(range(1,len(acumulacion_explicacion_varianza)+1), acumulacion_explicacion_varianza, marker='o')
plt.xlabel('Numero de componentes principales')
plt.ylabel('Varianza explicada acumulada (sumada)')
plt.title('Mi grafico para explicar una FUNCION que APLIQUE A LOS RESULTADO DEL PCA')
plt.show()

print("\n\n")

#Numero de componentes principales a CAPTURAR o RETENER
numeros_de_componentes_a_capturar= 2
pca= PCA(n_components= numeros_de_componentes_a_capturar)
pca_mercado= pca.fit_transform(ventas_mercado_scaled)

#Analisis e interpretacion de los componentes principales
componentes= pca.components_

# Visualizar la importancia de las características en el primer componente principal
plt.figure(figsize=(10, 6))
plt.bar(range(len(componentes[0])), componentes[0])
plt.xticks(range(len(componentes[0])),features_name, rotation=45,)
plt.xlabel('Características')
plt.ylabel('Importancia')
plt.title('Importancia de Características en el Primer Componente Principal')
plt.show()

print("\n\n")

# Obtener la lista de características con sus contribuciones a cada componente
feature_contributions=pd.DataFrame(componentes,columns=features_name)
print("Contribuciones de Características en Cada Componente Principal:")
print(feature_contributions.round(2))

print("\n En el primer componente principal (índice 0), las características que tienen contribuciones más altas son 'Tax 5%', 'Total', 'cogs', 'gross income' y 'Quantity'. Esto sugiere que estas características están altamente correlacionadas entre sí y son responsables de la variabilidad capturada por este componente principal. La característica 'Rating' tiene una contribución negativa, lo que indica una relación inversa con las otras características en este componente como se vio en la matriz de correlacion \n En el segundo componente principal (índice 1), la característica 'Unit price' tiene una contribución significativa, mientras que las otras características tienen contribuciones cercanas a cero. Esto sugiere que el segundo componente principal se enfoca en la variabilidad relacionada con 'Unit price' \n")

pca_grafico = pd.DataFrame({'PCA1':pca_mercado[:,0],'PCA2':pca_mercado[:,1],'Sucursal':data_frame_limpio['Branch']})
# Grafico de Barras de PCA1 y PCA2
plt.figure(figsize=(10,6))
plt.bar(x=['PCA1','PCA2'], height=pca.explained_variance_ratio_, color=['blue', 'green'])
plt.xlabel('Analisis de componentes principales')
plt.ylabel('Variancia acumulada')
plt.title('Contribucion de cada componente principal segun su varianza acumulada')
plt.show()

print("\n\n")

color_map = {'A': 'red', 'B': 'green', 'C': 'blue'}
pca_grafico['Color'] = pca_grafico['Sucursal'].map(color_map)
# Grafico de dispersion de PCA1 y PCA2 por Sucursal
plt.figure(figsize=(10, 6))
plt.scatter(x='PCA1', y='PCA2', c='Color', data=pca_grafico)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA1 vs PCA2 por Sucursal')
plt.show()

print("\n PCA se utiliza a menudo como una técnica de reducción de dimensionalidad o para encontrar patrones ocultos en los datos. \n PCA1 y PCA2 son las dos primeras componentes principales resultantes del análisis de componentes principales (PCA),conservando la mayor cantidad posible de variabilidad en los datos originales.\n Cada componente principal es una combinación lineal de las variables originales en el conjunto de datos. PCA1 y PCA2 son las dos combinaciones lineales más significativas y que capturan la mayor parte de la variabilidad en los datos. Estas componentes principales se eligen de manera que la primera componente (PCA1) explique la mayor cantidad posible de varianza en los datos, la segunda componente (PCA2) explique la siguiente mayor cantidad de varianza, y así sucesivamente.\n En el scatter plot de PCA1 vs PCA2 cada punto representa una observación en el conjunto de datos original. Los valores en estos ejes son las proyecciones de las observaciones en las direcciones de las dos primeras componentes principales.")
```
